```python
import random
import pandas as pd
import numpy as np

# definir los parámetros del algoritmo genético
tam_poblacion = 4
prob_mutacion = 0.8
generaciones = 50

# definir la capacidad de la mochila
lim_cal = 1200
lim_peso = 2

print(f"El límite de peso es: {lim_peso} kg y el límite de calorías es: {lim_cal} cal")

# definir la lista de pesos y calorías de los productos
cal = [500, 300, 250, 20, 200, 500, 180, 400]
peso = [0.7, 0.2, 0.8, 0.2, 0.6, 0.9, 0.6, 0.6]
lista_prod = pd.DataFrame ({"Cal":cal, "Peso":peso},index=["Coca-cola", "Pan", "Atún", "Agua", "Chocolate", "Papitas", "Fruta", "Palomitas"])
print(lista_prod)



c_1 = [0, 0, 0, 0]
p_1 = [0, 0, 0, 0]
hijo1 = [0, 0, 0, 0, 0, 0, 0, 0]
hijo2 = [0, 0, 0, 0, 0, 0, 0, 0]
hijo3 = [0, 0, 0, 0, 0, 0, 0, 0]
hijo4 = [0, 0, 0, 0, 0, 0, 0, 0]

# definir la función para calcular peso y calorías
def calcular(genes):
  total_peso = 0
  total_cal = 0
  for i in range(len(genes)):
    if genes[i] == 1:
      total_peso += peso[i]
      total_cal += cal[i]
  return total_peso, total_cal

# definir la función de cruce (crossover) al azar
def crossover(parent1, parent2):
  #point = random.randint(1, len(parent1)-1)
  point = 4
  child1 = parent1[:point] + parent2[point:]
  child2 = parent2[:point] + parent1[point:]
  return hijo1, hijo2

# definir la función de mutación
def mutation(individual, mutation_rate):
  for i in range(len(individual)):
    if random.random() < mutation_rate:
      individual[i] = 1 - individual[i]
  return individual


# crear una población inicial aleatoria
poblacion = [random.choices([0, 1], k=len(peso)) for i in range(tam_poblacion)]



for x in range(generaciones):

  #Función Fitness
  j = 0
  for genes in poblacion:
    p_1[j], c_1[j] = calcular(genes)
    j=j+1

  conejitos = pd.DataFrame ({"Genes":poblacion, "Cal":c_1, "Peso":p_1}, index=["Conejito1", "Conejito2", "Conejito3", "Conejito4"])

  def mejor(x):
    #(x[0] >= 1000) or ((abs(x[0]-1000)/1000)) and ((x[1] < 2) or (x[1] == 2))
    (x[0] >= lim_cal) and (x[1] <= lim_peso)
    return x

  #conejitos = conejitos.sort_values(by=['Cal', 'Peso'], key=mejor, ascending=False)
  conejitos = conejitos.sort_values(by=['Cal', 'Peso'], key=mejor)


  #Selección Fitness
  padre1 = conejitos.iloc[0, 0]
  padre2 = conejitos.iloc[1, 0]
  padre3 = conejitos.iloc[2, 0]

  #Cruzamiento de 1ro con 2do y 1ro con 3ro
  hijo1, hijo2 = crossover(padre1, padre2)
  hijo3, hijo4 = crossover(padre2, padre3)
  sig_generacion=[hijo1, hijo2, hijo3, hijo4]

  #Mutación
  for j in range(tam_poblacion):
    sig_generacion[j] = mutation(sig_generacion[j], prob_mutacion)

  #Actualización de la generación
  poblacion = sig_generacion

  #Aquí falta la trampa para ir guardando al conejito mejor de cada generación

print(conejitos)

```
